### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)

```python
conn = mysql.connector.connect(**config)

cursor = conn.cursor()
stat = "SELECT CASE_NUMBER FROM crimes WHERE DESCRIPTION  LIKE '%MINOR%';"

cursor.execute(stat)
rows = cursor.fetchall()
for i in rows:
    print("CASE NUMBER: ",i[0])
cursor.close()
conn.close()
```

### Problem 4

##### List all kidnapping crimes involving a child?

### Problem 6

### Problem 8

##### Which community area is most crime prone?

```python
conn = mysql.connector.connect(**config)

cursor = conn.cursor()
stat = '''
SELECT community_area_number, COUNT(*) as 'crime_count'
FROM crimes
GROUP BY community_area_number
ORDER BY crime_count DESC
LIMIT 1;
'''

cursor.execute(stat)
rows = cursor.fetchall()
for i in rows:
    print("Area num: ",i[0])
cursor.close()
conn.close()
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?

```python









conn = mysql.connector.connect(**config)

cursor = conn.cursor()
stat = '''
SELECT chic.COMMUNITY_AREA_NAME,COUNT(*) AS occ FROM crimes 
INNER JOIN chic ON crimes.COMMUNITY_AREA_NUMBER =chic.COMMUNITY_AREA_NUMBER  group by crimes.COMMUNITY_AREA_NUMBER, chic.COMMUNITY_AREA_NAME ORDER BY occ DESC;
'''

cursor.execute(stat)
rows = cursor.fetchall()
for i in rows:
    print("Name: ", i[0],  "Number : ", i[1])
cursor.close()
conn.close()

```
